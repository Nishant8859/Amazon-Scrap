```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
from IPython.display import display

```

```python
data = pd.DataFrame()
```

```python
base_url="https://www.amazon.in"
i=1
while i<=20:
    r = requests.get(base_url + "/s?k=bags&page=" + str(i) + "&crid=2M096C61O4MLT&qid=1653308124&sprefix=ba%2Caps%2C283&ref=sr_pg_1")
# Parsing the HTML
    print(i, r)
    if r.status_code == 503:
        time.sleep(1)
        continue
    else:
        
        soup = BeautifulSoup(r.content, 'html.parser')
        for div in soup.find_all('div',"sg-col sg-col-4-of-12 sg-col-8-of-16 sg-col-12-of-20 sg-col-12-of-24 s-list-col-right"):
            Product_Name = div.find("span", "a-size-medium a-color-base a-text-normal")
            Product_URL = div.find('a',"a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal",href=True)
            Product_Price = div.find('span',"a-price-whole")
            Rating = div.find('span',"a-icon-alt")
            Number_of_reviews = div.find('span',"a-size-base s-underline-text")
            
            d = {"Product URL":[Product_URL.get('href')],
            "Product Name":[Product_Name.text], 
            "Product Price":[Product_Price.text],
            "Rating":[Rating.text],
            "Number of reviews":[Number_of_reviews.text]}
            
            data = pd.concat([data, pd.DataFrame(d)])
        i+=1
        # for text in soup.find_all('span',"a-size-medium a-color-base a-text-normal"):
        #     Product_Name.append(text.text)
        # for link in soup.find_all('a',"a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"):
        #     Product_URL.append(text.text)
        # for text in soup.find_all('span',"a-price-whole"):
        #     Product_Price.append(text.text)
        # for text in soup.find_all('span',"a-icon-alt"):
        #     Rating.append(text.text)
        # for text in soup.find_all('span',"a-size-base s-underline-text"):
        #     Number_of_reviews.append(text.text)
```

```python
Product_URL_list = list(data["Product URL"])
base_url="https://www.amazon.in"
i=0
while i < len(Product_URL_list):
    r = requests.get(base_url + str(Product_URL_list[i]))
    print(i,r.status_code)
    if r.status_code == 503:
        time.sleep(1)
        continue
    else:
        soup = BeautifulSoup(r.content, 'html.parser')
        Description = soup.find_all('ul',"a-unordered-list a-vertical a-spacing-mini")
        Description_full = ""
        for j in Description:
            Description_full += j.text
        try:
            table = soup.find("div", "a-section table-padding")
            ASIN = table.find('td',"a-size-base prodDetAttrValue")
        except:
            #ASIN.string = None
            pass
        #print(ASIN.text)
        Product_Des = soup.find_all("div", "aplus-v2 desktop celwidget")
        paragraph = ""
        for para in Product_Des:
            Des = para.find_all('p', recursive=True)
            for p2 in Des:
                #print(p2.text, end="\n")
                paragraph += p2.text
        try:
            table2 = soup.find("div", "a-expander-content a-expander-section-content a-section-expander-inner")
            Manf = table2.find_all('td',"a-size-base prodDetAttrValue")
            #print(Manf[1].text)
        except:
            pass
        
        d2 = {"Description":[Description_full],
        "ASIN":[ASIN.text], 
        "Product Description":[paragraph],
        "Manufacturer":[Manf[1].text],}
        data2 = pd.concat([data2, pd.DataFrame(d2)])
        
        i+=1
    
```
